```julia
ticks = get_agg_trades(symbol, "2019-10-23T00:00", "2021-01-28T00:00")
t1 = 1
t2 = 800000
Close = ticks[t1:t2, "price"];

t = ticks[t1:t2, "trade_id"];

tss = ticks[t1:t2, "timestamp"];
```

```julia
lw = "1"
figure(figsize=(15, 5))
plot(t, Close, label=symbol, lw=lw)
plot(t, ID.ema(Close, n=10000), label="EMA1", lw=lw)
# plot(t, ID.ema(Close, n=20000), label="EMA2", lw=lw)
# plot(t, ID.ema(Close, n=30000), label="EMA3", lw=lw)
# plot(t, ID.ema(Close, n=40000), label="EMA4", lw=lw)
plot(t, ID.ema(Close, n=50000), label="EMA5", lw=lw)
# plot(t, ID.ema(Close, n=60000), label="EMA6", lw=lw)
# plot(t, ID.ema(Close, n=70000), label="EMA7", lw=lw)
# plot(t, ID.ema(Close, n=80000), label="EMA7", lw=lw)
# plot(t, ID.ema(Close, n=90000), label="EMA7", lw=lw)
plot(t, ID.ema(Close, n=100000), label="EMA10", lw=lw)
plot(t, ID.ema(Close, n=200000), label="EMA20", lw=lw)
legend(loc="best", frameon=false)
show()
```
